```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
print(netflix.groupby("Country").get_group("Japan"))
print("\n")
print(netflix.groupby("Country").get_group("India"))
print("\n")
print(netflix.groupby("Country").get_group("Canada"))
print("\n")
print(netflix.groupby("Country").get_group("United States"))
print("\n")
```

```python
NET= netflix.iloc[:10]
NET
```

## ploting 

```python
Genre=(NET["Genre"])
ReleaseYear=(NET["Release Year"])
# designing of the plot done in plt.plot
#plt.plot(days,drive_km,color="g",
        # linewidth=2,marker="o",markerfacecolor="r",
        # markersize=7)
plt.xlabel("Shows and Movies",fontsize=20)
plt.ylabel("Release Year",fontsize=20)
plt.title("Netflix Movie data",fontsize=25)
plt.xticks(rotation=45,fontsize=10)
plt.plot(Genre,ReleaseYear, marker="o",color="red",markerfacecolor="k",linewidth=3,markersize=15)	
plt.show()
```

```python
plt.figure(figsize=(10,5))
Genre=(NET["Genre"])
ReleaseYear=(NET["Release Year"])
plt.xlabel("Shows and Movies",fontsize=20)
plt.ylabel("Release Year",fontsize=20)
plt.xticks(rotation=45,fontsize=10)
plt.title("Netflix Movie data",fontsize=25)
plt.bar(Genre,ReleaseYear,color="red",alpha=0.4)
```

```python
plt.figure(figsize=(10,5))
Genre=NET['Genre']
ReleaseYear=NET['Release Year']
c=["y","b","m","g","black","grey","orange","r","m","g","gold"]
plt.bar(Genre,ReleaseYear,color=c,edgecolor='r',label="Genre")
plt.title("movie shows",fontsize=18)
plt.xlabel("movies")
plt.ylabel("release year")
plt.xticks(rotation=45,fontsize=10)
plt.legend()
plt.show()
```

```python
Genre=(NET["Genre"])
ReleaseYear=(NET["Release Year"])
# designing of the plot done in plt.plot
plt.plot(Genre,ReleaseYear,color="red",linewidth=1,
         marker="*",markerfacecolor="cyan",markersize=15,
         linestyle='--')#dashed ,dotted ,dashdot
plt.xlabel("Genre",fontsize=20)
plt.ylabel("ReleaseYear",fontsize=20)
plt.title("netflix movie data",fontsize=25)
plt.show()
```
